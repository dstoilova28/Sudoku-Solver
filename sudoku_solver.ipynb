<h1 style="text-align: center;">Sudoku Solver using Backtracking</h1>

Sudoku is a logic puzzle consisting of a 9x9 grid divided into nine 3x3 subgrids.

The goal is to fill in the empty cells so that:

- In each row, the numbers from 1 to 9 appear exactly once.
- In each column, the numbers from 1 to 9 appear exactly once.
- In each 3x3 subgrid, the numbers from 1 to 9 appear exactly once.

In this project, we implement a Sudoku solver using a recursive backtracking algorithm.

```python
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

def print_board(board):
    for row in board:
        print(row)

print_board(board)

```

## Step 1: Board Representation

The Sudoku board is represented as a 9x9 two-dimensional list in Python.

- Each inner list represents one row.
- The value `0` represents an empty cell.
- A cell can be accessed using `board[row][col]`.

This structure allows us to treat Sudoku as a matrix problem.

## Step 2: Printing the Board

The function `print_board()` iterates through each row of the board.

- `for row in board` loops through the rows.
- `print(row)` displays each row.

This helps us visualize the current state of the puzzle.


```python
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]
def find_empty(board):
    for i in range(9):          
        for j in range(9):      
            if board[i][j] == 0:
                return (i, j)  
    return None

print(find_empty(board))

```

## Step 3: Finding an Empty Cell

The function scans the board to locate a cell containing `0`.

- If an empty cell is found, its position `(row, col)` is returned.
- If no empty cells remain, the function returns `None`.

This step tells the algorithm where to continue solving.

```python
def is_valid(board, num, row, col):
    
    if num in board[row]:
        return False

    
    for i in range(9):
        if board[i][col] == num:
            return False

    
    start_row = (row // 3) * 3
    start_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if board[start_row + i][start_col + j] == num:
                return False

    return True

print(is_valid(board, 5, 0, 2))  
print(is_valid(board, 2, 0, 2))  

```

## Step 4: Validity Check

Before placing a number, we verify that:

- The number is not already in the same row.
- The number is not in the same column.
- The number is not in the corresponding 3x3 subgrid.

This guarantees that the Sudoku rules are respected.

```python
def solve(board):
    empty = find_empty(board)
    if not empty:
        return True 
    else:
        row, col = empty

    for num in range(1, 10):
        if is_valid(board, num, row, col):
            board[row][col] = num  

            if solve(board):
                return True  

            board[row][col] = 0  

    return False  


    print("Начална дъска:")
print_board(board)

if solve(board):
    print("\nРешена дъска:")
    print_board(board)
else:
    print("Няма решение!")

```

## Step 5: Backtracking Algorithm

The `solve()` function uses recursion and backtracking:

1. Find an empty cell.
2. Try numbers from 1 to 9.
3. If a number is valid, place it.
4. Recursively continue solving.
5. If a dead end is reached, reset the cell and try another number.

This process continues until the board is completely solved.

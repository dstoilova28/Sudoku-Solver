```python
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

def print_board(board):
    for row in board:
        print(row)

print_board(board)

```

```python
def find_empty(board):
    for i in range(9):          # редове
        for j in range(9):      # колони
            if board[i][j] == 0:
                return (i, j)   # връща позицията
    return None

print(find_empty(board))

```

```python
def is_valid(board, num, row, col):
    # Проверка на ред
    if num in board[row]:
        return False

    # Проверка на колона
    for i in range(9):
        if board[i][col] == num:
            return False

    # Проверка на 3x3 квадрат
    start_row = (row // 3) * 3
    start_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if board[start_row + i][start_col + j] == num:
                return False

    return True

print(is_valid(board, 5, 0, 2))  # False, защото 5 вече е в реда
print(is_valid(board, 2, 0, 2))  # True, може да се сложи

```

```python
def solve(board):
    empty = find_empty(board)
    if not empty:
        return True  # няма празни клетки → решението е готово
    else:
        row, col = empty

    for num in range(1, 10):
        if is_valid(board, num, row, col):
            board[row][col] = num  # пробваме числото

            if solve(board):
                return True  # ако решението работи, спираме

            board[row][col] = 0  # иначе връщаме назад (backtrack)

    return False  # ако нищо не работи, връщаме False


    print("Начална дъска:")
print_board(board)

if solve(board):
    print("\nРешена дъска:")
    print_board(board)
else:
    print("Няма решение!")

```
